```python
import pandas as pd
```

```python
# previsores
X = pd.read_csv('entrada_numerica.csv')
X
```

```python
# classe
y = pd.read_csv('saida.csv')
y
```

```python
#Usado para dividir o conjunto de treinanemto e teste
import sklearn
from sklearn.model_selection import train_test_split
```

```python
# Vai dividir em 75% treinamento e 25% teste
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(X, y, test_size=0.25)
```

```python
X_treinamento.shape, y_treinamento.shape
```

```python
X_teste.shape,y_teste.shape
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
```

```python
rede_neural = Sequential([
    tf.keras.layers.Input(shape=(5,)),
    tf.keras.layers.Dense(units=16, activation='relu', kernel_initializer='random_uniform'),
    tf.keras.layers.Dense(units=1, activation='sigmoid'),
])
```

```python
#resumo dos parametros
rede_neural.summary()
```

```python
otimizador = tf.keras.optimizers.Adam(learning_rate=0.001, decay=0.0001, clipvalue=0.5)
```

```python
#Isso garante que o modelo use o Adam com os parâmetros definidos durante o treinamento.
rede_neural.compile(optimizer='adam', loss='binary_crossentropy', metrics=['binary_accuracy'])
```

```python
# treinamento da rede neural, o batch_size irá controlar os pesos e analisar em lote, então ele minimiza o erro
# Epochs irá passar pelo modelo de treinamento 30x
rede_neural.fit(X_treinamento, y_treinamento, batch_size=10, epochs=30)
```

```python
# realizando previsoes com o modelo treinado, se as previsoes forem menor ou igual a 0.5, será 0 ou False, se for maior será 1 ou True
previsoes = rede_neural.predict(X_teste)
previsoes = (previsoes > 0.5)
previsoes
```

```python
y_teste
```

```python
# confusion_matrix compara as previsoes feitas pelo modelo com as reais(Verdadeiros)
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
# Probabilidade de acerto ficou entre 80%
accuracy_score(y_teste, previsoes)
```

```python
#resultado final: Perda de 62% (com uma variação)
#                 acuracia de 80%
rede_neural.evaluate(X_teste, y_teste)
```
