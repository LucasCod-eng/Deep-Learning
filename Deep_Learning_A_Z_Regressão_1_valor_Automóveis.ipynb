Gerando as primeiras previsões

```python
regressor = Sequential([
    tf.keras.layers.InputLayer(shape=(316,)),
    tf.keras.layers.Dense(units=158, activation='relu'),
    tf.keras.layers.Dense(units=158, activation='relu'),
    tf.keras.layers.Dense(units=1, activation='linear'),
])
```

```python
regressor.summary()
```

```python
#A
#real: 12.000
#previsao: 13.000

#B
#real: 11.500
#previsao: 10.000
#
regressor.compile(optimizer='adam', loss='mean_absolute_error', metrics=['mean_absolute_error'])
```

```python
regressor.fit(X, y, epochs=100, batch_size=300)
```
